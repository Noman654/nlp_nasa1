```python
data_df = []
with open('abc.txt') as f:
    lines = f.readlines()
    for line in tqdm(lines):
        category = str(line).strip()
        url = f"https://ntrs.nasa.gov/api/citations/search??subjectCategory={category}"
        data = requests.get(url)
        # break
        if data.status_code != 200:
            print(category)
            break
        else:
            d = pd.json_normalize(data.json()['results'])
            data_df.append(d)
```

```python
data_df_a = pd.concat(data_df)
```
